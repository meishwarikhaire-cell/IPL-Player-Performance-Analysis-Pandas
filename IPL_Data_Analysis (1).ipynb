```python
#Task 1
# Filter death overs
death_overs = deliveries[(deliveries['over'] >= 16) & (deliveries['over'] <= 20)]

# Group by batsman
batsman_stats = death_overs.groupby('batsman').agg(
    total_runs=('batsman_runs','sum'),
    balls_faced=('batsman_runs','count')
).reset_index()

# Calculate strike rate
batsman_stats['strike_rate'] = (batsman_stats['total_runs'] / batsman_stats['balls_faced']) * 100

# Filter batsmen with at least 200 balls
qualified_batsmen = batsman_stats[batsman_stats['balls_faced'] >= 200]

# Top 10 batsmen by strike rate
top10 = qualified_batsmen.sort_values(by='strike_rate', ascending=False).head(10)

top10
```

```python
merged_df = pd.merge(deliveries, matches[['id','Season']],
                     left_on='match_id', right_on='id',
                     how='left')
```

```python
season_runs = merged_df.groupby(['Season','batsman'])['batsman_runs'].sum().reset_index()
```

```python
season_runs = season_runs.sort_values(['Season','batsman_runs'], ascending=[True, False])
```

```python
orange_cap = season_runs.drop_duplicates(subset='Season', keep='first')

orange_cap[['Season','batsman','batsman_runs']]
```
